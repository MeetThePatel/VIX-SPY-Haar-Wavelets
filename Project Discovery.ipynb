```python
# DONE
from alpha_vantage.timeseries import TimeSeries
import pandas
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

```python
# DONE
ts = TimeSeries(key='V0MIN5N4S35ZE668', output_format='pandas')
```

```python
#DONE
def getDailyCloseArray(ticker):
    data, metadata = ts.get_daily(ticker)
    return data['4. close']

#DONE
vix_daily = getDailyCloseArray("^VIX")
vix_daily.plot()
plt.title("Plot of ^VIX")

#Done
trendvecVIX, detailvecVIX = HaarWaveletTransform(vix_daily)
m2 = list(range(int(len(vix_daily)/2)))
X2 = [2*i for i in list(m2)]
plt.plot(X2, detailvecVIX)

#DONE
spy_daily = getDailyCloseArray("SPY")
spy_daily.plot()
plt.title("Plot of $SPY")

# DONE
def HaarWaveletTransform(timeseries):
    m = list(range(int(len(timeseries)/2)))
    T1 = [(timeseries[2*i] + timeseries[2*i + 1])/2 for i in m]
    D1 = [(timeseries[2*i] - timeseries[2*i + 1])/2 for i in m]
    return T1, D1

#DONE
trendvec, detailvec = HaarWaveletTransform(spy_daily)
m = list(range(int(len(spy_daily)/2)))
X1 = [2*i for i in list(m)]

plt.plot(X1, detailvec)
plt.plot(X2, detailvecVIX)
plt.plot(X2, np.zeros(len(detailvec)))
plt.show()

spy_int = integrate.cumtrapz(detailvec, X1, initial=0)
plt.plot(X1, spy_int)

vix_int = integrate.cumtrapz(detailvecVIX, X1, initial=0)
plt.plot(X1, vix_int)

diff_list = [x1 + x2 for (x1, x2) in zip(detailvec, detailvecVIX)]
diff_int = integrate.cumtrapz(diff_list, X1, initial=0)
plt.plot(X1, diff_int)

```
